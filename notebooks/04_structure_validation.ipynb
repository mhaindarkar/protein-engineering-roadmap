# 04 Structure Validation

## Goal
Evaluate AlphaFold / ColabFold predictions.

## Purpose
- load prediction JSON files
- extract pLDDT and pTM
- summarize structure-confidence metrics


```python
import json
import glob
import pandas as pd

records = []

for file in glob.glob("../results/alphafold/*.json"):
    with open(file) as f:
        data = json.load(f)

    records.append({
        "file": file,
        "plddt": sum(data["plddt"]) / len(data["plddt"]),
        "ptm": data.get("ptm", None)
    })

metrics_df = pd.DataFrame(records)
metrics_df.head()
```
