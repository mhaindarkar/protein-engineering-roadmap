# 04 Structure Validation

## Goal
Evaluate AlphaFold / ColabFold predictions.

## Purpose
- load prediction JSON files
- extract pLDDT and pTM
- summarize structure-confidence metrics


***Extract metrics (same as before)***

```python
import pandas as pd
import json
import glob

# Load design table
design_df = pd.read_csv("../results/tables/design_candidates.csv")

# Load AlphaFold metrics
records = []

for file in glob.glob("../results/alphafold/*.json"):
    with open(file) as f:
        data = json.load(f)

    records.append({
        "file": file,
        "plddt": sum(data["plddt"]) / len(data["plddt"]),
        "ptm": data.get("ptm", None)
    })

metrics_df = pd.DataFrame(records)

# Create IDs
metrics_df["id"] = metrics_df["file"].str.extract(r"(seq_\d+)")
design_df["id"] = [f"seq_{i}" for i in range(len(design_df))]

# Merge
combined_df = design_df.merge(metrics_df, on="id")

# Save it
combined_df.to_csv("../results/tables/design_validation.csv", index=False)

combined_df
```
